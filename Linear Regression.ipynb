```python
from sklearn.datasets import load_diabetes
data = load_diabetes()
```

```python
data
```

```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df=load_diabetes()
```

```python
dataset = pd.DataFrame(df.data)
```

```python
dataset.columns=df.feature_names
```

```python
dataset.head()
```

```python
# INDEPENDENT AND DEPENDENT FEATURES
X=dataset
y=df.target
```

```python
y
```

```python
#TRAIN TEST SPLIT
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest=train_test_split(X,y,test_size=0.3,random_state=42)
```

```python
# STANDARISING THE DATSET
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
```

```python
xtrain= scaler.fit_transform(xtrain)
```

```python
xtrain
```

```python
xtest=scaler.transform(xtest)
```

```python
from sklearn.linear_model import LinearRegression
#cross validation
from sklearn.model_selection import cross_val_score
```

```python
lr=LinearRegression()
```

```python
lr.fit(xtrain,ytrain)
```

```python
mse= cross_val_score(lr,xtrain,ytrain,scoring='neg_mean_squared_error',cv=10)
```

```python
np.mean(mse)
```

```python
#PREDICTION
ypred = lr.predict(xtest)
```

```python
ypred
```

```python
import seaborn as sns
sns.distplot(ypred-ytest)
```

```python
from sklearn.metrics import r2_score
```

```python
score = r2_score(ypred,ytest)
```

```python
score
```

```python
# RIDGE REGRESSION ALGORITHM
from sklearn.linear_model import Ridge
from sklearn.model_selection import GridSearchCV
```

```python
ridge=Ridge()
```

```python
ridge
```

```python
parameters={'alpha':[10,12,15,20,22,25,30,35]}
ridgecv = GridSearchCV(ridge,parameters,scoring='neg_mean_squared_error',cv=5)
ridgecv.fit(xtrain,ytrain)
```

```python
print(ridgecv.best_params_)
```

```python
print(ridgecv.best_score_)
```

```python
ridge_pred = ridgecv.predict(xtest)
```

```python
sns.distplot(ridge_pred-ytest)
```

```python
score=r2_score(ridge_pred,ytest)
```

```python
score
```

```python
# LASSO REGRESSION
from sklearn.linear_model import Lasso
from sklearn.model_selection import GridSearchCV
```

```python
lasso=Lasso()
```

```python
lasso
```

```python
parameters={'alpha':[10,12,15,18,20,25,30,35]}
lasso_cv= GridSearchCV(lasso,parameters,scoring='neg_mean_squared_error',cv=5)
lasso_cv.fit(xtrain,ytrain)
```

```python
print(lasso_cv.best_params_)
```

```python
print(lasso_cv.best_score_)
```

```python
lasso_pred=lasso_cv.predict(xtest)
```

```python
print(r2_score(lasso_pred,ytest))
```

```python
sns.distplot(lasso_pred-ytest)
```

```python
sns.kdeplot(lasso_pred - ytest, fill=True)
```
